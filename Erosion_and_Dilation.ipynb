```python
# Import necessary libraries
import cv2
import os
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

# Clone the dataset repository
!git clone https://github.com/jestineeun/OpenCV_Basics_MEXE4102_Adolfo_Francisco.git

# Define the dataset folder
dataset_folder = "/content/OpenCV_Basics_MEXE4102_Adolfo_Francisco/Dataset/Testing/"


# Process all images in the folder
for filename in os.listdir(dataset_folder):
    file_path = os.path.join(dataset_folder, filename)

    # Check file extension to ensure it's an image
    if filename.endswith(('.jpg', '.png', '.jpeg')):

# Load the image
        image = cv2.imread(file_path)

# Display the resized image
        resized_image = cv2.resize(image, (500, 500))

# Display the canny image
        gray = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)
        canny_image = cv2.Canny(gray, 150, 200)

# Define the kernel for erosion and dilation
        kernel = np.ones((1, 2), np.uint8)

# Apply Erosion
        eroded_image = cv2.erode(canny_image, kernel, iterations=1)

# Apply Dilation
        dilated_image = cv2.dilate(canny_image, kernel, iterations=1)

# Apply Erosion after Dilation
        final_image = cv2.erode(dilated_image, kernel, iterations=1)

# Display intermediate steps and final results
        plt.figure(figsize=(15, 10))

# Original Image

        plt.subplot(1, 5, 1)
        plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
        plt.title("Original Image")
        plt.axis("off")

# Canny Image
        plt.subplot(1, 5, 2)
        plt.imshow(canny_image, cmap="gray")
        plt.title("Canny Image")
        plt.axis("off")

# Eroded Image
        plt.subplot(1, 5, 3)
        plt.imshow(eroded_image, cmap="gray")
        plt.title("Eroded Image")
        plt.axis("off")

# Dilated Image
        plt.subplot(1, 5, 4)
        plt.imshow(dilated_image, cmap="gray")
        plt.title("Dilated Image")
        plt.axis("off")

# Final Image
        plt.subplot(1, 5, 5)
        plt.imshow(final_image, cmap="gray")
        plt.title("Final Image")
        plt.axis("off")


        plt.show()

```
